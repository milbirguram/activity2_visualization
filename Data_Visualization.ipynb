#DataVis with Seaborn

Assumptions - In the magazine, I did not consider the skin images that were in B/W. The three classes are size of skin samples, color of skin samples as required and third one is intensity that has value from 1-5 to take camera flash into consideration during the magazine photoshoots.

The Data Visualization has primarily used bar graphs, histograph and pie chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv("dataset.csv")
```

```python
train.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Color', data=train)
```

```python
sns.countplot(x='Size', data=train)
```

```python
sns.countplot(x='Intensity', data=train)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Size', hue='Color', data=train, palette='RdBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Intensity', hue='Color', data=train, palette='RdBu_r')
```

```python
sns.distplot(train['Intensity'].dropna(), kde=False, color='darkred', bins=40)
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Color', y='Intensity', data=train, palette='winter')
```

```python
sns.heatmap(train.corr())
```

```python
sns.barplot(x='Color', y='Intensity', data=train)
```

```python
# also showing the piechart of percentage 
labels = 'Fair', 'Brown', 'Dark'
sizes = [79, 31, 7]
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
